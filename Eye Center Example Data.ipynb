```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from Google Drive
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/Cataract_Surgery_Procurement_Dataset.csv'
dataset = pd.read_csv(file_path)


# Inventory Analysis - Identify low stock items and reorder needs
reorder_data = dataset[dataset['Current_Stock_Level'] <= dataset['Reorder_Level']]
print("Items that need reordering (Current Stock <= Reorder Level):")
print(reorder_data[['Product_Name', 'Current_Stock_Level', 'Reorder_Level']])

# Visualize the low stock items
plt.figure(figsize=(10, 6))
plt.barh(reorder_data['Product_Name'], reorder_data['Current_Stock_Level'], color='salmon')
plt.axvline(x=reorder_data['Reorder_Level'].mean(), color='gray', linestyle='--', label='Average Reorder Level')
plt.title('Low Stock Items Needing Reorder')
plt.xlabel('Current Stock Level')
plt.ylabel('Product Name')
plt.legend()
plt.show()

# Cost Analysis - Overall and per category
print("\nCost Summary:")
print(dataset['Unit_Cost'].describe())

# Total spending per category
category_costs = dataset.groupby('Product_Type')['Total_Cost'].sum().sort_values(ascending=False)
print("\nTotal Spending per Category:")
print(category_costs)

# Visualize spending per category
plt.figure(figsize=(10, 6))
category_costs.plot(kind='bar', color='lightblue')
plt.title('Total Spending by Product Category')
plt.xlabel('Product Type')
plt.ylabel('Total Cost ($)')
plt.show()

# Supplier Performance - Average lead time by supplier
supplier_lead_time = dataset.groupby('Supplier_Name')['Lead_Time (Days)'].mean().sort_values()
print("\nAverage Lead Time by Supplier:")
print(supplier_lead_time)

# Visualize supplier performance (lead times)
plt.figure(figsize=(10, 6))
supplier_lead_time.plot(kind='bar', color='green')
plt.title('Average Lead Time by Supplier')
plt.xlabel('Supplier Name')
plt.ylabel('Lead Time (Days)')
plt.show()

# Demand Forecasting - Identify frequently ordered items
# Count frequency of each product in the dataset
order_counts = dataset['Product_Name'].value_counts().sort_values(ascending=False)
print("\nFrequently Ordered Products:")
print(order_counts.head(10))

# Visualize top 10 most frequently ordered items
plt.figure(figsize=(10, 6))
order_counts.head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Most Frequently Ordered Products')
plt.xlabel('Product Name')
plt.ylabel('Order Frequency')
plt.show()

# Summary of Key Insights for Management
print("\nSummary of Key Insights:")
print("1. Items that are low in stock and need reordering have been identified, which management can prioritize.")
print("2. Cost analysis reveals the overall spending per category, with the most expenditure on certain high-cost categories.")
print("3. Supplier performance has been evaluated, providing insights into which suppliers have the fastest lead times.")
print("4. Frequently ordered products have been identified, helping to inform future ordering and inventory planning decisions.")



# Stock Turnover Rate - Calculate how quickly items are used up
# Assuming Quantity_Ordered represents usage, calculate the stock turnover rate
dataset['Stock_Turnover_Rate'] = dataset['Quantity_Ordered'] / dataset['Current_Stock_Level']
turnover_data = dataset[['Product_Name', 'Stock_Turnover_Rate']].sort_values(by='Stock_Turnover_Rate', ascending=False)
print("\nStock Turnover Rate (Higher means quicker turnover):")
print(turnover_data.head(10))

# Visualize top 10 products by Stock Turnover Rate
plt.figure(figsize=(10, 6))
turnover_data.head(10).plot(kind='barh', x='Product_Name', y='Stock_Turnover_Rate', color='teal')
plt.title('Top 10 Products by Stock Turnover Rate')
plt.xlabel('Stock Turnover Rate')
plt.ylabel('Product Name')
plt.show()

# Supplier Cost Comparison - Average Unit Cost per Supplier for each Product Type
supplier_cost_comparison = dataset.groupby(['Supplier_Name', 'Product_Type'])['Unit_Cost'].mean().unstack()
print("\nAverage Unit Cost per Supplier for each Product Type:")
print(supplier_cost_comparison)

# Heatmap for supplier cost comparison by product type
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(supplier_cost_comparison, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Average Unit Cost per Supplier for each Product Type')
plt.xlabel('Product Type')
plt.ylabel('Supplier Name')
plt.show()

# Monthly Spend Analysis
# Convert Order_Date to datetime format and extract month and year
dataset['Order_Date'] = pd.to_datetime(dataset['Order_Date'])
dataset['Order_Month'] = dataset['Order_Date'].dt.to_period('M')
monthly_spend = dataset.groupby('Order_Month')['Total_Cost'].sum()
print("\nMonthly Spend Analysis:")
print(monthly_spend)

# Plot monthly spending pattern
plt.figure(figsize=(12, 6))
monthly_spend.plot(kind='line', marker='o', color='darkblue')
plt.title('Monthly Spend Analysis')
plt.xlabel('Month')
plt.ylabel('Total Cost ($)')
plt.grid(True)
plt.show()

# Forecasted Reorder Needs based on Stock Turnover Rate
# Products with high turnover and low stock should be prioritized for reordering
forecasted_reorder_data = dataset[(dataset['Stock_Turnover_Rate'] > 1) & (dataset['Current_Stock_Level'] <= dataset['Reorder_Level'])]
print("\nForecasted Reorder Needs (High turnover and low stock):")
print(forecasted_reorder_data[['Product_Name', 'Current_Stock_Level', 'Reorder_Level', 'Stock_Turnover_Rate']])

# ABC Analysis for Inventory Prioritization
# Calculate cumulative cost contribution
total_cost = dataset['Total_Cost'].sum()
dataset['Cost_Contribution'] = (dataset['Total_Cost'] / total_cost) * 100
dataset['Cumulative_Cost_Contribution'] = dataset['Cost_Contribution'].cumsum()

# Assign ABC categories
dataset['ABC_Category'] = pd.cut(dataset['Cumulative_Cost_Contribution'], bins=[0, 70, 90, 100], labels=['A', 'B', 'C'])
abc_summary = dataset.groupby('ABC_Category').agg({'Product_Name': 'count', 'Total_Cost': 'sum'})
print("\nABC Analysis Summary:")
print(abc_summary)

# Visualize ABC analysis results
plt.figure(figsize=(8, 5))
abc_summary['Total_Cost'].plot(kind='pie', autopct='%1.1f%%', labels=['A', 'B', 'C'], colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('ABC Analysis - Cost Contribution by Category')
plt.ylabel('')
plt.show()

# Usage Variability - Coefficient of Variation in Quantity Ordered for each Product
usage_variability = dataset.groupby('Product_Name')['Quantity_Ordered'].std() / dataset.groupby('Product_Name')['Quantity_Ordered'].mean()
usage_variability = usage_variability.dropna().sort_values(ascending=False)
print("\nTop 10 Products by Usage Variability (High Variability indicates unpredictable demand):")
print(usage_variability.head(10))

# Visualize Usage Variability
plt.figure(figsize=(10, 6))
usage_variability.head(10).plot(kind='barh', color='coral')
plt.title('Top 10 Products by Usage Variability')
plt.xlabel('Coefficient of Variation')
plt.ylabel('Product Name')
plt.show()




# IOL Break down


import pandas as pd
import matplotlib.pyplot as plt


# Filter for IOL products only
iol_data = dataset[dataset['Product_Type'] == 'Intraocular Lens']

# Calculate total spend by IOL type
spend_by_iol = iol_data.groupby('Product_Name')['Total_Cost'].sum().sort_values(ascending=False)

# Plot as a bar chart
plt.figure(figsize=(10, 6))
spend_by_iol.plot(kind='bar', color='skyblue')
plt.title('Total Spend by IOL Type')
plt.xlabel('IOL Type')
plt.ylabel('Total Spend ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Calculate total spend by IOL type again if needed
spend_by_iol = iol_data.groupby('Product_Name')['Total_Cost'].sum()

# Plot as a donut chart
plt.figure(figsize=(8, 8))
plt.pie(spend_by_iol, labels=spend_by_iol.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.title('Total Spend by IOL Type')
plt.tight_layout()
plt.show()


# Create a DataFrame for the treemap chart
treemap_data = iol_data.groupby('Product_Name')['Total_Cost'].sum().reset_index()

# Plot the treemap
fig = px.treemap(
    treemap_data,
    path=['Product_Name'],
    values='Total_Cost',
    color='Total_Cost',
    color_continuous_scale='YlOrRd',
    title='Total Spend by IOL Type'
)
fig.update_layout(margin=dict(t=40, l=0, r=0, b=0))
fig.show()




# Convert Order_Date to datetime format
iol_data['Order_Date'] = pd.to_datetime(iol_data['Order_Date'])
iol_data['Month'] = iol_data['Order_Date'].dt.to_period('M')

# Calculate monthly spend on IOLs
monthly_spend = iol_data.groupby('Month')['Total_Cost'].sum()
monthly_spend.plot(kind='line', marker='o', figsize=(10, 6), title='Monthly Spend on IOLs')
plt.xlabel('Month')
plt.ylabel('Total Spend ($)')
plt.grid()
plt.show()


# Calculate total spend by supplier
spend_by_supplier = iol_data.groupby('Supplier_Name')['Total_Cost'].sum().sort_values(ascending=False)

# Plot the spend by supplier
spend_by_supplier.plot(kind='bar', color='teal', figsize=(10, 6))
plt.title('Total Spend by Supplier')
plt.xlabel('Supplier')
plt.ylabel('Total Spend ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Calculate cumulative cost contribution
iol_data['Cumulative_Cost'] = iol_data['Total_Cost'].cumsum() / iol_data['Total_Cost'].sum()

# Apply ABC categorization
iol_data['ABC_Category'] = pd.cut(iol_data['Cumulative_Cost'], bins=[0, 0.7, 0.9, 1], labels=['A', 'B', 'C'])

# Summarize the ABC categorization
abc_summary = iol_data.groupby('ABC_Category')['Total_Cost'].sum()
abc_summary.plot(kind='pie', autopct='%1.1f%%', startangle=140, title='ABC Analysis of IOL Spend')
plt.ylabel('')
plt.show()

# Identify items that need reordering
reorder_items = iol_data[iol_data['Current_Stock_Level'] <= iol_data['Reorder_Level']]
print("Items that need reordering:")
print(reorder_items[['Product_Name', 'Current_Stock_Level', 'Reorder_Level']])






```
